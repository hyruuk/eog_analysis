# Quantifying eye movements from EOG recordings

Eye movements are a major source of artifacts during M/EEG experiments. This is especially relevant in the context of naturalistic viewing paradigms because unconstrained eye movements might differ between experimental conditions, especially when we compare viewing of different video content. 

In my case, I want to compare brain networks dynamics during eyes-open resting state, viewing of abstract movie and viewing of naturalistic movie (i.e. involving complex scenes, characters and social interactions).
Because we wanted to limit eye-movements during the experiment without degrading too much the viewing conditions, we asked ourselves if we should present the video at full size or reduced. To decide between these options, we presented the videos twice to our pilot subject, once at full size and once at a size reduced by 50%. That way, I can use the EOG data to see if presenting the video at 50% size led to reductions in eye-movements quantity and amplitude. If not, we will keep presenting at full size because it's more comfortable for the subject. If yes, we might decide to present at 50% or run more analysis to see if the eye-movements are problematic for the testing of our research hypothesis.


**_As a quick note, I realize that I don't have a good criterion that will help me decide if I should go for 50% or 100% videos. That's what makes these analysis "exploratory"._ **

Whatever.


**Here are some important factors to consider, that are specific to our experimental setup :**

- Distance to screen : 100cm
- Screen width : 43cm
- Screen height : 33cm


**Here are the different kinds of eye-movements we might want to quantify :**

- Blinks : the eyelids closes for a split of a seconde, generating a huge stereotyped potential mostly visible on vertical EOG
- Saccades : both eyes roll quickly in one direction on another in order to fixate a new location. Most saccades are not solely vertical nor horizontal, so they should be seen on both vEOG and hEOG)
- Smooth poursuit eye movements (SPEM): these are the continuous movements performed to track a moving object. They are usually more difficult to identify because they produce less stereotyped activity. They are very important in the context of naturalistic viewing but we might be forced to drop their analysis if I don't find a good approach to identify them from EOG (which looks difficult if not impossible).

## Using univariate convolutional sparse coding (CSC) to detect "atoms" (stereotyped patterns of activity) and their occurence through the signal.

Since I had no idea on how to approach the problem, I went to see my friend [Mainak Jas](https://jasmainak.github.io/), who advised me **against** using ICA to compare the weight of eye-related components in differents conditions (and so, from different decompositions). The reason being that ICA is a non-convex model, which seems to mean that there is no unique solution for a decomposition, and starting the process with different random seeds might lead to different results (according to his words).

He suggested however that I use a new tool he contributed developping, to see if it could be of any help for my questions. This tool is called [alphaCSC](https://alphacsc.github.io/) and it looks very promising for many neuroscientific applications. I have to admit that I don't understand much of it's working for now, but I'm on it.
Here are the few things I got from now :

- It's based on a machine learning technique called sparse dictionnary learning
- It works by learning "atoms", or prototypical events that are reproduced across the time series
- When the atoms are worked out, the technique allows to decompose the signal into a sum of "activation functions" (?) that describe the temporal occurence of the different atoms.

So, if I'm correct, we could try to decompose our signal, identify the atoms relating to blinks and saccades, and then compare the activation functions of both recordings. If the activation function of recording A shows more events than that of recording B for the "blink atom", I could then say that there were more blinks during the recording A.

**One thing that should be noted here is that Mainak advised me to concatenate the signals from the two conditions in order to learn atoms that would be shared between the recordings. If I don't do it that way, the activation functions I will want to compare would have different meaning, and thus couldn't be compared.**

### So, here is my first try :

```python
import numpy as np
from numpy import hamming
from alphacsc import learn_d_z
from mne.filter import filter_data

reg = 6.0
n_times = 2500
n_times_atom = 350
n_trials = 100
n_atoms = 3
n_iter = 60
random_state = 10
sfreq = 1200

# Let's just take the code from alphaCSC's tutorials and use our data instead.
X = np.concatenate((EOG_100[0], EOG_50[0]))
#X = filter_data(X.astype(np.float64), sfreq, l_freq=1, h_freq=None, fir_design='firwin')
#print(X)
#0/0

# First we need to segment the signal. I'm not sure why it's necessary
overlap = 0
starts = np.arange(0, X.shape[0] - n_times, n_times - overlap)
stops = np.arange(n_times, X.shape[0], n_times - overlap)
X_new = []
for idx, (start, stop) in enumerate(zip(starts, stops)):
    if idx >= n_trials:
        break
    X_new.append(X[start:stop])
X_new = np.vstack(X_new)
del X

# Then we normalize the signal
X_new -= np.mean(X_new)
X_new /= np.std(X_new)

# CSC is subject to edge effects, so this hamming window should help us prevent it
X_new *= hamming(n_times)[None, :]

pobj, times, d_hat, z_hat, reg = learn_d_z(X_new, n_atoms, n_times_atom,
                                           reg=reg, n_iter=n_iter,
                                           random_state=random_state, n_jobs=1)
```

```python
import numpy as np
from scipy.io import loadmat

import matplotlib.pyplot as plt

from mne.filter import filter_data
from alphacsc import GreedyCDL

n_atoms = 2
reg = 0.5
n_iter = 60
n_times_atom = 400
random_state = 42

#X = loadmat('EOG_50.mat')['EOG_50']

#X = X[0, 50000:100000][None, :]  # crop signal to demo

X = np.concatenate((EOG_100[0], EOG_50[0])).reshape(1, EOG_100[0].shape[0] + EOG_50[0].shape[0])
print(X.shape)

# filter data
sfreq = 200.
X = filter_data(X, sfreq, l_freq=1.0, h_freq=40.)

# z-score to have sane reg parameter
X -= np.mean(X)
X /= np.std(X)

cdl = GreedyCDL(
    # Shape of the dictionary
    n_atoms=n_atoms,
    n_times_atom=n_times_atom,
    # Request a rank1 dictionary with unit norm temporal and spatial maps
    rank1=False,
    uv_constraint='joint',
    # apply a temporal window reparametrization
    window=True,
    # at the end, refit the activations with fixed support and no reg to unbias
    unbiased_z_hat=True,
    # Initialize the dictionary with random chunk from the data
    D_init='chunk',
    # rescale the regularization parameter to be a percentage of lambda_max
    lmbd_max="scaled",
    reg=reg,
    # Number of iteration for the alternate minimization and cvg threshold
    n_iter=100,
    eps=1e-6,
    # solver for the z-step
    solver_z="lgcd",
    solver_z_kwargs={'tol': 1e-3,
                     'max_iter': 100000},
    # solver for the d-step
    solver_d='alternate_adaptive',
    solver_d_kwargs={'max_iter': 300},
    # sort atoms by explained variances
    sort_atoms=True,
    # Technical parameters
    verbose=1,
    random_state=0,
    n_jobs=1)

cdl.fit(X[None, ...])
```

```python
plt.plot(cdl.D_hat_[0][0])
plt.figure()
plt.plot(cdl.z_hat_[0][0], '-o')
plt.plot(X.T, alpha=0.5)
plt.show()
```

# Detecting EOG events with MNE-Python

The MNE-Python toolbox conveniently provides a function that can identify EOG events, including blinks and maybe some of the biggest saccades. From what I understood, it works by filtering the EOG channels and then detecting peaks in the signal.
Let's see if that function detects more events for the 100% than for the 50% condition :

```python
from mne.preprocessing import find_eog_events

raw_100 = read_raw_ctf(fname_100, preload=True, verbose=False)
raw_50 = read_raw_ctf(fname_50, preload=True, verbose=False)
```

```python
eog_events_100 = find_eog_events(raw_100, ch_name='EEG057', verbose=False)
eog_events_50 = find_eog_events(raw_50, ch_name='EEG057', verbose=False)
print('Number of events in 100% condition : ' + str(len(eog_events_100)))
print('Number of events in 50% condition : ' + str(len(eog_events_50)))
```

Ok so, looks like there are clearly more events in the 100% condition... But what is the impact of these events on the MEG data ?

To answer that question, we can segment the data around the events, and average them to produce ERFs. If the ERFs are strong, we can assume that the signal will be very contamined by occurence of a blink.

```python
from hytools.meg_utils import get_ch_pos, array_topoplot

baseline=(None, 0.0)
tmin=-0.2
tmax=0.8

event_id = 998
epochs_100 = mne.Epochs(raw_100, events=eog_events_100[:80], event_id=event_id, tmin=tmin,
                    tmax=tmax, baseline=baseline, reject=None, picks=None, preload=True)
epochs_50 = mne.Epochs(raw_50, events=eog_events_50, event_id=event_id, tmin=tmin,
                    tmax=tmax, baseline=baseline, reject=None, picks=None, preload=True)

evoked_100 = epochs_100.average()
evoked_50 = epochs_50.average()

evoked_100.plot();
evoked_50.plot();
```
